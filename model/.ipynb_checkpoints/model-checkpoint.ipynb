```python
from pathlib import Path
import random
import numpy as np 
import matplotlib.pyplot as plt
from PIL import Image
import os
from tqdm.auto import tqdm
import time
import pandas as pd

import torch
from torch.utils.data import DataLoader
from torchvision import datasets, transforms, models
import torch.nn as nn
import torchinfo
```

```python
images_path = Path("data_demo")

train_dir = images_path / "train"
test_dir = images_path / "test"
train_dir, test_dir
```

```python
random_image_path = random.choice(list(images_path.glob("*/*/*.jpg")))
image_label = random_image_path.parent.stem

image = Image.open(random_image_path)
image_np = np.array(image)

plt.imshow(image_np)
plt.title(f"Image label: {image_label}, shape: {image_np.shape}")
plt.axis("off")
```

```python
weights_1 = models.EfficientNet_B0_Weights.DEFAULT
auto_transforms = weights_1.transforms()
auto_transforms
```

```python
train_data = datasets.ImageFolder(root=train_dir,
                                   transform=auto_transforms)
test_data = datasets.ImageFolder(root=test_dir,
                                 transform=auto_transforms)
train_data, test_data 
```

```python
class_labels = train_data.classes
class_labels
```

```python
len(train_data), len(test_data)
```

```python
image, label = train_data[0][0], train_data[0][1]
print(f"Tensor: {image}")
print(f"Label: {label}")
print(f"shape: {image.shape}")
```

```python
plt.imshow(image.permute(1, 2, 0)) # Torch have [color_channels, height, width], numpy [height, width, colot_channels]
plt.axis("off")
plt.title(class_labels[label])
```

```python
BATCH_SIZE = 32
NUM_WORKERS = os.cpu_count()

torch.manual_seed(42)

train_dataloader = DataLoader(train_data, 
                               batch_size=BATCH_SIZE, 
                               shuffle=True,
                               num_workers=NUM_WORKERS)

test_dataloader = DataLoader(test_data, 
                              batch_size=BATCH_SIZE, 
                              shuffle=True,
                              num_workers=NUM_WORKERS)
```

```python
model_1 = models.efficientnet_b0(weights=weights_1)
```

```python
torchinfo.summary(model=model_1,
        input_size=(32, 3, 224, 224),
        col_names = ["input_size", "output_size", "num_params", "trainable"],
        col_width=20,
        row_settings=["var_names"]    
)
```

```python
for param in model_1.features.parameters():
    param.requires_grad = False
```

```python
output_shape = len(class_labels)

model_1.classifier = torch.nn.Sequential(
    torch.nn.Dropout(p=0.2, inplace=True),
    torch.nn.Linear(in_features=1280,
                    out_features=output_shape,
                    bias=True)
)
```

```python
torchinfo.summary(model=model_1,
        input_size=(32, 3, 224, 224),
        col_names = ["input_size", "output_size", "num_params", "trainable"],
        col_width=20,
        row_settings=["var_names"]    
)
```

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model_1.parameters(), lr=0.001)
```

```python
def train(model, train_dataloader, test_dataloader, optimizer, loss_fn, epochs):
    results= {"train_loss": [],
                  "train_acc": [],
                  "test_loss": [],
                  "test_acc": [],
        }
    
    for epoch in tqdm(range(epochs)):
        #1
        model.train()
        train_loss = 0
        train_acc = 0
        test_loss = 0 
        test_acc = 0

        for batch, (X_train, y_train) in enumerate(train_dataloader):
            y_pred = model(X_train)
            loss = loss_fn(y_pred, y_train)
            train_loss += loss.item()
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            y_pred_labels = torch.argmax(torch.softmax(y_pred, dim=1), dim=1)
            train_acc += ((y_pred_labels == y_train).sum().item() / len(y_pred)) * 100
    
        train_loss = train_loss / len(train_dataloader)
        train_acc = train_acc / len(train_dataloader)
        #2

        model.eval()
        with torch.inference_mode():
            for batch, (X_test, y_test) in enumerate(test_dataloader):
                test_pred = model(X_test)
                loss = loss_fn(test_pred, y_test)
                test_loss += loss.item()
                test_pred_labels = torch.argmax(torch.softmax(test_pred, dim=1), dim=1)
                test_acc += ((test_pred_labels == y_test).sum().item() / len(y_test)) * 100
        
            test_loss = test_loss / len(test_dataloader)
            test_acc = test_acc / len(test_dataloader)

        print(f"Epoch: {epoch+1} | Train loss: {train_loss:.4f}, Train acc: {train_acc:.2f} % | Test loss: {test_loss:.4f}, Train acc: {train_acc:.2f} % ")
        
        results["train_loss"].append(train_loss)
        results["train_acc"].append(train_acc)
        results["test_loss"].append(test_loss)
        results["test_acc"].append(test_acc)
                
    return results    
        
        
```

```python
start = time.time()
results_model_1 = train(model=model_1,
                        train_dataloader=train_dataloader,
                        test_dataloader=test_dataloader,
                        loss_fn=loss_fn,
                        optimizer=optimizer,
                        epochs=10)
end = time.time()

print(f"Training time of model_1: {end-start:.4f} seconds.")
```

```python
def plot_curves(results, model_name):
    epochs = len(results["train_loss"])
    plt.figure(figsize=(14,7))
    plt.suptitle(model_name)
    
    plt.subplot(2,4,1)
    plt.plot(range(0,epochs), results["train_loss"],)
    plt.title("Train loss")
    plt.xlabel("epochs")
    
    plt.subplot(2,4,2)
    plt.plot(range(0,epochs), results["test_loss"],)
    plt.title("Test loss")
    plt.xlabel("epochs")

    plt.subplot(2,4,3)
    plt.plot(range(0,epochs), results["train_acc"],)
    plt.title("Train accuracy [%]")
    plt.xlabel("epochs")
    
    plt.subplot(2,4,4)
    plt.plot(range(0,epochs), results["test_acc"],)
    plt.title("Test accuracy [%]")
    plt.xlabel("epochs")
    
    plt.show
```

```python
plot_curves(results_model_1, "EfficientNet")
plot_curves(results_model_2, "VGG16")
plot_curves(results_model_3, "ResNet")
```
